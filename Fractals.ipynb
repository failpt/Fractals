## Що таке фрактали?

Фрактал — це математичний об'єкт, який характеризується повторюваністю своїх елементів на різних масштабах і відрізняється від звичайних геометричних форм за рахунок своєї фрактальної розмірності, фрактал постає однаковим при будь-якому збільшенні.

Фрактали стали не лише предметом математичнмх досліджень, але й знайшли застосування в різних наукових галузях. Наприклад, вони використовуються у фізиці, біології та у комп'ютерній графіці для візуалізації складних структур, таких як ландшафти. Особливо цікавим є їх вплив на моделювання природних явищ, де фрактальні структури часто спостерігаються, як-от у формах сніжинок.

У минулому математика мала тенденцію ігнорувати об'єкти, які не вкладалися у рамки класичного числення. Такі об'єкти вважалися "патологічними" і не заслуговували на детальне вивчення. Проте, з часом з'ясувалося, що негладкі, нерегулярні множини можуть бути джерелом цінних знань, особливо у розумінні природних феноменів. Фрактальна геометрія відкрила нові шляхи для аналізу таких структур.

Мандельброт спочатку визначив фрактал як множину з розмірністю Хаусдорфа-Безиковича, яка перевищує топологічну розмірність. Однак пізніше він переглянув це визначення, роблячи акцент на самоподібності фракталів. Таке оновлене визначення краще описує багатство фрактальних структур, які можна зустріти в природі та наукових дослідженнях, включаючи багато фракталів, що зустрічаються у фізиці.

## Найвідоміші фрактали

### Множина Кантора
Множина Кантора — один із найвідоміших фракталів.
Створення множини Кантора можна представити як такий процес: розглянемо відрізок $[0,1]$. На першому кроці видаляємо відкритий відрізок $(\frac{1}{3},\frac{2}{3})$, отримуючи два замкнених відрізки: $[0,\frac{1}{3}]$ та $[\frac{2}{3},1]$. На кожному наступному кроці розділяємо кожний існуючий відрізок на три рівні частини і видаляємо середню. Процес продовжується нескінченно.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, IntSlider, FloatSlider, Dropdown, Checkbox, HBox

def draw_cantor_set(start, length, depth, y, ax):
    ax.plot([start, start + length], [y, y], color='#000000', lw=2.5)
    if depth == 0:
        return
    
    draw_cantor_set(start, length / 3, depth-1, 1 - depth * 0.1, ax)
    draw_cantor_set(start + length * 2 / 3, length / 3, depth-1, 1 - depth * 0.1, ax)

def cantor_set(iterations):
    fig, ax = plt.subplots(figsize=(44, 22))
    
    draw_cantor_set(0, 1, iterations, 1 - iterations * 0.1, ax)
        
    ax.invert_yaxis()
    ax.set_aspect(aspect='auto')
    ax.axis('off')
    plt.show()

interactive(cantor_set, iterations=IntSlider(min=0, max=6, step=1, value=4, description="Ітерації"))
```

### Трикутник Серпінського
Нехай $T_{0}$ буде рівностороннім трикутником. Для кожного трикутника, який ми отримаємо після $n$ ітерацій, позначимо його як $T_{n}$. Тоді трикутник Серпінського на $n$-й ітерації можна описати так: 
$$T_{n+1}=F(T_{n})$$
де $F$ – це функція, яка бере трикутник $T_{n}$, розділяє його на чотири менших трикутники і видаляє центральний. 

Формально, якщо $T_{n}$ має вершини $A_{n}$, $B_{n}$, $C_{n}$, то отримана фігура буде складатися з трьох трикутників з вершинами $A_{n}$, $M_{A_{n}B_{n}}$, $M_{A_{n}C_{n}}$, $B_{n}$, $M_{A_{n}B_{n}}$, $M_{C_{n}B_{n}}$ та $C_{n}$, $M_{C_{n}B_{n}}$, $M_{A_{n}C_{n}}$, де $M_{XY}$ – це середина відрізка $XY$.
Після нескінченного числа ітерацій ми отримаємо фрактал Серпінського, який можна описати як множину всіх точок, що належать хоча б одному з трикутників $T_{n}$ для всіх $n$.

```python
def draw_sierpinski(depth, ax, vertices):
    if depth == 0:
        ax.fill(*zip(*vertices), '#6d00fc')
    else:
        p1 = vertices[0]
        p2 = vertices[1]
        p3 = vertices[2]
        mid12 = (p1 + p2) / 2
        mid23 = (p2 + p3) / 2
        mid31 = (p3 + p1) / 2
        
        draw_sierpinski(depth - 1, ax, [p1, mid12, mid31])
        draw_sierpinski(depth - 1, ax, [mid12, p2, mid23])
        draw_sierpinski(depth - 1, ax, [mid31, mid23, p3])

def sierpinski(depth): 
    fig, ax = plt.subplots(figsize=(7.6, 6.7))
    vertices = np.array([[0.5, np.sqrt(3)/2], [0, 0], [1, 0]])
    draw_sierpinski(depth, ax, vertices)
    ax.axis('off')
    plt.show()
    
interactive(sierpinski, depth=IntSlider(min=0, max=7, step=1, value=5, description="Ітерації"))
```

### Множина Мандельброта
Множина Мандельброта визначається множиною комплексних чисел $c$, для яких послідовність $z_n$ (де $z_{0}=0$ та $z_{n+1}=z_{n}^2+c$) може або залишатися обмеженою або прямувати до нескінченності зі збільшенням $n$. Якщо послідовність залишається обмеженою (не дійшовши до нескінченності) для конкретного $c$ після великої кількості ітерацій, ми вважаємо, що $c$ належить множині Мандельброта. В іншому випадку, якщо вона дійшла до нескінченності в певний момент ітерації, $c$ не вважається частиною множини Мандельброта.
Ітераційна функція: $$z_{n+1}=z_{n}^2+c$$ Тут $z$ – це комплексне число, що на початку ітерацій $z_{0}=0$.

```python
def mandelbrot(h, w, x_min, x_max, y_min, y_max, max_iter, z_pow):
    x = np.linspace(x_min, x_max, num=w).reshape((1, w))
    y = np.linspace(y_min, y_max, num=h).reshape((h, 1))
    C = np.tile(x, (h, 1)) + 1j * np.tile(y, (1, w))
    
    Z = np.zeros(C.shape, dtype=complex)
    M = np.zeros(C.shape, dtype=int)
    for i in range(max_iter):
        mask = np.abs(Z) < 2
        M[mask] = i
        Z[mask] = Z[mask] ** z_pow + C[mask]
    return M

def mandelbrot_zoom(re_center, im_center, zoom, max_iter, cm, z_pow):   
    size = 400
    scale_factor = 1.5 / zoom
    x_min = re_center - scale_factor
    x_max = re_center + scale_factor
    y_min = im_center - scale_factor
    y_max = im_center + scale_factor
    
    M = mandelbrot(size, size, x_min, x_max, y_min, y_max, max_iter, z_pow)
    
    fig, ax = plt.subplots(figsize=(7.8, 7.8))   
    plt.imshow(M, extent=(x_min, x_max, y_min, y_max), cmap=cm)
    colorbar = plt.colorbar()
    plt.rcParams.update({'axes.titlesize': 17, 'axes.labelsize': 17})
    plt.gca().set_xlabel('Re')
    plt.gca().set_ylabel('Im')
    plt.show()

interactive(
    mandelbrot_zoom,
    re_center=FloatSlider(min=-2.5, max=1.5, step=0.05, value=-0.75, description="Re"),
    im_center=FloatSlider(min=-2, max=2, step=0.05, value=0, description="Im"),
    zoom=FloatSlider(min=1, max=100, step=1, value=1, readout_format='.1f', description="zoom"),
    max_iter=IntSlider(min=10, max=200, step=1, value=50, description='iter'),
    cm=Dropdown(options=['turbo', 'magma', 'viridis', 'cividis', 'hot'], value='magma', description='Палітра'),
    z_pow=FloatSlider(min=1.0, max=10.0, step=0.05, value=2, description='z^p')
)
```

#### Ремарка для користувача
Слайдери `Re` і `Im` визначають координати центру візуалізації множини Мандельброта, `zoom` регулює рівень збільшення чи зменшення фракталу, а `iter` контролює максимальну кількість ітерацій, яка впливає на деталізацію та якість зображення фракталу.

Шкала відображає кількість ітерацій, які потрібно для того, щоб точка вийшла за межі визначеного радіусу (в даному випадку радіус $2$).
Якщо в процесі ітерації модуль $z$ стає більшим за $2$, то точка вважається "біжучою" і не належить множині Мандельброта. Для вивчення динаміки системи на практиці використовують максимальне число ітерацій(`iter`). Якщо після встановленого числа ітерацій $z$ не стає більшим за $2$, точка вважається належною до множини Мандельброта.
Різні кольори допомагають легко ідентифікувати області, де точки вийшли за радіус після різної кількості ітерацій. Таким чином, ця шкала відображає "швидкість" збіжності або розбіжності точок у фракталі Мандельброта.

Під час дослідження множин Мандельброта та Жуліа мене зацікавив їх вигляд з формулою $z_{n+1}=z_{n}^p+c$, де p буде відмінним від двійки, тому змінюючи значення на слайдері `z^p` можна дослідити зміну форми множини Мандельброта за зміни степіню z.

**Палітри viridis та cividis створені для дальтоніків**

### Множина Жуліа
Множина Жуліа — це множина комплексних чисел, яка визначається системою ітерацій: $z_{n+1}=z_{n}^2+c$, де z – комплексне число, а $c$ – деяка комплексна константа.

```python
def julia_set(h, w, x_min, x_max, y_min, y_max, c, max_iter, z_pow):
    x = np.linspace(x_min, x_max, num=w).reshape((1, w))
    y = np.linspace(y_min, y_max, num=h).reshape((h, 1))
    Z = np.tile(x, (h, 1)) + 1j * np.tile(y, (1, w))

    M = np.zeros(Z.shape, dtype=int)
    for i in range(max_iter):
        mask = np.abs(Z) < 2
        M[mask] = i
        Z[mask] = Z[mask] ** z_pow + c
    return M

def julia_zoom(re_center, im_center, zoom, c_real, c_imag, max_iter, cm, z_pow):   
    size = 400
    scale_factor = 1.5 / zoom
    x_min = re_center - scale_factor
    x_max = re_center + scale_factor
    y_min = im_center - scale_factor
    y_max = im_center + scale_factor
    
    J = julia_set(size, size, x_min, x_max, y_min, y_max, complex(c_real, c_imag), max_iter, z_pow)
    
    fig, ax = plt.subplots(figsize=(7.8, 7.8))
    plt.imshow(J, extent=(x_min, x_max, y_min, y_max), cmap=cm)
    colorbar = plt.colorbar()
    plt.rcParams.update({'axes.titlesize': 17, 'axes.labelsize': 17})
    plt.gca().set_xlabel('Re')
    plt.gca().set_ylabel('Im')
    plt.title(f"c = {c_real:.2f} + {c_imag:.2f}i")
    plt.show()

interactive(
        julia_zoom,
        re_center=FloatSlider(min=-2, max=2, step=0.05, value=0, description="Re"),
        im_center=FloatSlider(min=-2, max=2, step=0.05, value=0, description="Im"),
        zoom=FloatSlider(min=1, max=100, step=1, value=1, readout_format='.1f', description="zoom"),
        c_real=FloatSlider(min=-2, max=2, step=0.01, value=-0.86, description='c_real'),
        c_imag=FloatSlider(min=-2, max=2, step=0.01, value=0.12, description='c_imag'),
        max_iter=IntSlider(min=10, max=200, step=1, value=50, description='iter'), 
        cm = Dropdown(options=['turbo', 'magma', 'viridis', 'cividis', 'hot'], value='magma', description='Палітра'),
        z_pow = FloatSlider(min=1.0, max=10.0, step=0.05, value=2, description='z^p')
    )
```

Відмінність між множиною Мандельброта і множиною Жуліа полягає в початковій точці ітераційного процесу та константі $c$. Для множини Мандельброта початкова точка завжди дорівнює $0$, а $c$ змінюється. Для множини Жуліа $c$ є постійною, а початкова точка $z$ змінюється.
#### Ремарка для користувача
Для множини Жуліа повзунки `Re` і `Im` так само визначають положення центру візуалізації, `zoom` регулює рівень збільшення, `z^p` відповідає за степінь $z$ у формулі $z_{n+1}=z_{n}^p+c$, a `c_real` із `c_imag` змінюють дійсну та уявну частини комплексного параметру $c$, який визначає форму множини Жуліа, і `iter` контролює глибину ітерацій, впливаючи на деталізацію зображення.

Як і в попередній візуалізації кольорова шкала показує, як швидко значення в певній точці виходять за межі радіуса $2$ під час ітераційного процесу. Колір кожної точки відповідає кількості ітерацій, необхідних для того, щоб визначити, чи належить точка до множини чи ні.

## Фрактали в природі

Як було зазначено раніше — фрактали відомі як геометричні фігури, які демонструють подібність у всьому діапазоні масштабу. І вони, насправді, всюдисущі в природі. У 1960-1970-х роках Мандельброт зрозумів, що багато явищ реального світу — хмари, сніжинки, берегові лінії — мають подібні властивості. Вони демонструють «самоподібність», шаблони, які повторюються у все менших і менших масштабах та мають нечіткі межі. У своїй роботі я розглядаю декілька прикладів таких фракталів.

### Сніжинки
Немає й двох абсолютно однакових по формі сніжинок, але багато з них представляють фрактали, оскільки гілки сніжинки породжують власні бічні гілки.

Фрактальний візерунок сніжинки, відомий як **сніжинка Коха**, став одним з найперших та найвідоміших фракталів, вперше описаним шведським математиком Хельге фон Кохом.

Сніжинка Коха утворюється шляхом застосування процесу генерації Кривої Коха до кожної сторони рівностороннього трикутника. Крок ітерації полягає в тому, що ми беремо кожен відрізок та ділимо його на три рівні частини. В середній третині відрізка ми замінюємо його двома сторонами рівностороннього трикутника, основою якого є видалена середня третина. Таким чином, кожен відрізок перетворюється на чотири відрізки. 

```python
def koch_snowflake(order):
    def koch_curve(points, order):
        if order == 0:
            return points

        new_points = []
        for i in range(len(points) - 1):
            p1, p2 = points[i], points[i + 1]
            u = (p2 - p1) / 3
            p3 = p1 + u
            p4 = p3 - u * np.exp(np.pi * 1j / 1.5) 
            p5 = p1 + 2 * u
            new_points.extend([p1, p3, p4, p5])
        new_points.append(points[-1])
        return koch_curve(new_points, order - 1)

    triangle = np.exp(np.pi * 1j * np.arange(3) / 1.5)
    snowflake = koch_curve(triangle.tolist() + [triangle[0]], order)
    return np.real(snowflake), np.imag(snowflake)

def plot_koch_snowflake(order): 
    x, y = koch_snowflake(order)
    fig, ax = plt.subplots(figsize=(7.8, 7.8))
    ax.plot(x, y, lw=0.5, color='#1a4c7d')
    ax.fill(x, y, color='#edf7f7')
    ax.set_axis_off()
    ax.set_aspect('equal')
    plt.show()

interactive(plot_koch_snowflake, order=IntSlider(min=0, max=6, step=1, value=2, description="Ітерації"))
```

На $n$-му кроці утворена фігура буде мати $3 \cdot 4^n$ сторін.

### Дерева
Модель фрактального дерева, показана нище, використовується для імітації природних форм дерев та кущів, демонструючи, як із простих правил можуть виникати складні структури.

Візуалізація розпочинається із заданої початкової точки та довжини стовбура: Від точки початку малюється початкова лінія (стебло дерева) з певною довжиною, яка зменшується з кожною ітерацією (у моєму випадку в 0.7 разів). На кінці стебла (новій точці початку) сворюються дві нові гілки, нахилені під заданим кутом. Такий процес продовжується до тих пір, поки не буде досягнуто задану глибину ітерацій.

```python
def draw_fractal_tree(origin, angle, length, depth, branch_angle, symmetry, tree_data, ll_data):
    end_point = origin + length * np.array([np.cos(angle), np.sin(angle)])
    
    if depth == 1 or depth == 2:
        ll_data['x'].extend([origin[0], end_point[0], None])
        ll_data['y'].extend([origin[1], end_point[1], None])
        if depth == 1:
            return
    else:
        tree_data['x'].extend([origin[0], end_point[0], None])
        tree_data['y'].extend([origin[1], end_point[1], None]) 

    new_length = length * 0.7
    new_depth = depth - 1

    draw_fractal_tree(end_point, angle + branch_angle/symmetry, new_length, new_depth, branch_angle, symmetry, tree_data, ll_data)
    draw_fractal_tree(end_point, angle - branch_angle, new_length, new_depth, branch_angle, symmetry, tree_data, ll_data)

def plot_fractal_tree(iterations, branch_angle, symmetry):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal')
    ax.axis('off')

    origin = np.array([0, 0])
    angle = np.pi / 2
    length = 0.2
    depth = iterations 

    tree_data = {'x': [], 'y': []}
    ll_data = {'x': [], 'y': []}
    draw_fractal_tree(origin, angle, length, depth, branch_angle, symmetry, tree_data, ll_data)

    ax.plot(tree_data['x'], tree_data['y'], color='#573705', lw=0.7)
    ax.plot(ll_data['x'], ll_data['y'], color='#7ca803', lw=0.3)

    plt.show()

interactive(plot_fractal_tree,
            iterations=IntSlider(min=1, max=17, step=1, value=6, description='Ітерації'),
            branch_angle=FloatSlider(min=0.3, max=np.pi/2.3, step=0.05, value=0.55, description='Нахил'),
            symmetry=FloatSlider(min=0.3, max=3, step=0.05, value=1.5, description='Симетрія'))
#, readout=False
```

Початкова точка `origin`$ = (x_{0}, y_{0})$ і кінцева точка `end_point` кожної гілки визначаються через їх координати на площині:$$\text{end_point}_{x} = x_{0} + \text{length} \cdot \cos(\text{angle})$$ $$\text{end_point}_{y} = y_{0} + \text{length} \cdot \sin(\text{angle})$$

Нахил гілок `branch_angle` та симетрія `symmetry` контролюють кути нахилу при генерації нових гілок: 
- $\text{новий кут нахилу лівої гілки} = \text{angle} + \frac{\text{branch_angle}}{\text{symmetry}}$
- $\text{новий кут нахилу правої гілки} = \text{angle} - \text{branch_angle}$

### Папороть Барнслі
Фрактал папороті Барнслі є математичною моделлю, що імітує природну форму папороті. Ця модель використовує ітеративний процес для створення складної структури, яка дивовижно нагадує зелені листки папороті.

Нехай $F$ - функція, що трансформує точку $(x,y)$ в нову на основі обраної випадковим чином підфункції $f_1,f_2,f_3,$ або $f_4$. Кожна з цих функцій відповідає за певну трансформацію частини фрактала:
- $f_1(P)=\begin{bmatrix}
t_1 & 0 \\
0 & 0.16
\end{bmatrix}P + \begin{bmatrix}
0 \\
0
\end{bmatrix}$   відповідає за стебло (бере будь-яку точку та переносить її в лінію в центрі в залежності від паремтру $t_1$)
- $f_2(P) = \begin{bmatrix}
0.85 & t_2 \\
-0.04 & 0.85
\end{bmatrix}P + \begin{bmatrix}
0 \\
1.6
\end{bmatrix}$   рухає точки вверх, та залежачи від параметру $t_2$ в сторону
- $f_3(P) = \begin{bmatrix}
0.2 & -0.26 \\
0.23 & 0.22
\end{bmatrix}P + \begin{bmatrix}
t_3 \\
1.6
\end{bmatrix}
$   та  $f_4(P) = \begin{bmatrix}
-0.15 & 0.28 \\
0.26 & 0.24
\end{bmatrix}P + \begin{bmatrix}
t_4 \\
0.44
\end{bmatrix}$   повертають та переносять всі точки в право та в ліво в залежності від параметрів $t_3$ та $t_4$ для створення окремих листків

На початку, маючи точку $(0,0)$, ми застосовуємо одну з цих траснформацій випадковим чином, маючи задану ймовірність вибору кожної функції. Функція $f_2$ викликається з ймовірністю 85%, що робить її основною трансформацією для візуалізації папоротня. 

```python
import random

def f1(x, y, t):
    return t*x, 0.16*y

def f2(x, y, t):
    return 0.85*x + t*y, -0.04*x + 0.85*y + 1.6

def f3(x, y, t):
    return 0.2*x - 0.26*y + t, 0.23*x + 0.22*y + 1.6

def f4(x, y, t):
    return -0.15*x + 0.28*y + t, 0.26*x + 0.24*y + 0.44

def generate_points(n, t1, t2, t3, t4):
    x, y = 0, 0
    points = []
    for _ in range(n):
        r = random.random()
        if r < 0.01:
            x, y = f1(x, y, t1)
        elif r < 0.85:
            x, y = f2(x, y, t2)
        elif r < 0.93:
            x, y = f3(x, y, t3)
        else:
            x, y = f4(x, y, t4)
        points.append((x, y))
    return points

def plot_fern(t1, t2, t3, t4, ifax):
    n = 300000
    points = generate_points(n, t1, t2, t3, t4)
    x, y = zip(*points)
    plt.figure(figsize=(7, 8))
    plt.plot(x, y, ',', color='#00821a')
    plt.rcParams.update({'axes.labelsize': 17})
    if not ifax:
        plt.axis('off')
        
    plt.show()

interactive(plot_fern,
            t1=FloatSlider(value=0.00, min=-0.5, max=0.5, step=0.05, description='t1'),
            t2=FloatSlider(value=0.04, min=-0.05, max=0.1, step=0.01, description='t2'),
            t3=FloatSlider(value=0.00, min=-0.7, max=0.7, step=0.05, description='t3'),
            t4=FloatSlider(value=0.00, min=-0.7, max=0.7, step=0.05, description='t4'), 
            ifax=Checkbox(value=False, description="Вісі")
           )
```

Слайдери над візуалізацією дають можливість дослідити поведінку зображення залежачи від раніше зазначених параметрів $t_1$, $t_2$, $t_3$ та $t_4$.

#### Ремарка для користувача
Вище наведені функції є афінними перетвореннями. Афінне перетворення — це математична операція, яка дозволяє виконувати різноманітні геометричні трансформації, такі як масштабування, поворот, віддзеркалення та зсув. Воно включає множення вектора позиції об'єкта на матрицю та додавання вектора перенесення. Загальна форма афінного перетворення для двовимірної точки $P$ може бути записана так:$$f(P) = A \cdot P + B$$
де $A$ — це матриця перетворення, $P$ — вектор позиції, а $B$ — вектор перенесення. 

Або так:
$$f(P) = \begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix} \begin{bmatrix}
x \\
y
\end{bmatrix} + \begin{bmatrix}
b_1 \\
b_2
\end{bmatrix}$$
Якщо розглянути це за координатами, результатом буде:$$f(P)_{x} = a_{11} \cdot x + a_{12} \cdot y + b_{1}$$ $$f(P)_{y} = a_{21} \cdot x + a_{22} \cdot y + b_{2}$$
де $f(P)_{x}$ та $f(P)_{y}$ — це координати $x$ та  $y$ нової точки відповідно після афінного перетворення.

### Алгоритм "Diamond-Square"
Алгоритм "Diamond-Square" або алгоритм "зміщення середньої точки" являє собою метод створення фрактальних поверхонь, що імітують висотні рельєфи природних ландшафтів. Ідея полягає в рекурсивному поділенні площини на менші сегменти та додаванні випадкового зміщення до висот кожної точки, що створює нерегулярність поверхні, схожу на природний рельєф.

В основі алгоритму для генерації 3D зображень лежить двовимірна сітка, кожен елемент якої представляє висоту (додаткове значення, яке можна вважати третьою координатою) відповідної точки на ландшафті. Процес починається з ініціалізації чотирьох кутових точок сітки випадковими висотами, які слугують "зернами" для подальшої генерації рельєфу.

Алгоритм складається з двох фундаментальних кроків, які повторюються на кожній ітерації:
1. *Крок Діаманта(Diamond step):* На цьому етапі для кожного квадрату визначається центральна точка, висота якої встановлюється як середнє значення висот чотирьох кутових точок плюс випадкова компонента, що створює ефект "горбів" та "ям".
2. *Крок квадрата:* Далі, для кожного "діаманта" (утвореного на попередньому кроці) визначаються нові точки на серединах сторін, висоти яких також визначаються через середнє значення сусідніх точок із випадковою компонентою.

З кожною ітерацією масштаб випадкових змін зменшується, що дозволяє створювати більш дрібні деталі ландшафту та досягати ефекту природної ерозії. В результаті, ландшафт стає більш реалістичним із меншими змінами висоти на малих відстанях.

```python
from mpl_toolkits.mplot3d import Axes3D

def diamond_square(size, roughness, seed=9):
    grid = np.zeros((size, size))
    np.random.seed(seed)
    grid[0, 0] = np.random.rand()
    grid[0, size - 1] = np.random.rand()
    grid[size - 1, 0] = np.random.rand()
    grid[size - 1, size - 1] = np.random.rand()

    step_size = size - 1
    while step_size > 1:
        half_step = step_size // 2

        for x in range(0, size - 1, step_size):
            for y in range(0, size - 1, step_size):
                avg = (grid[x, y] + grid[x + step_size, y] + grid[x, y + step_size] + grid[x + step_size, y + step_size]) / 4
                grid[x + half_step, y + half_step] = avg + (np.random.rand() - 0.5) * roughness

        for x in range(0, size, half_step):
            for y in range((x + half_step) % step_size, size, step_size):
                avg = sum([
                    grid[(x - half_step) % (size - 1), y],
                    grid[(x + half_step) % (size - 1), y],
                    grid[x, (y + half_step) % (size - 1)],
                    grid[x, (y - half_step) % (size - 1)]
                ]) / 4
                grid[x, y] = avg + (np.random.rand() - 0.5) * roughness

        step_size //= 2
        roughness /= 2

    return grid

def plot_fractal_landscape(power, roughness, cm):
    size = 2 ** power + 1
    seed = sum(ord(char) for char in cm) % 100
    
    landscape = diamond_square(size, roughness, seed)
    fig = plt.figure(figsize=(8.5, 8.5))
    ax = fig.add_subplot(111, projection='3d')
    x = np.arange(size)
    y = np.arange(size)
    x, y = np.meshgrid(x, y)
    ax.plot_surface(x, y, landscape, rstride=1, cstride=1, cmap=cm)
    
    for axis in [ax.xaxis, ax.yaxis, ax.zaxis]:
        axis.pane.fill = False
        axis.pane.set_edgecolor('white')
    
    ax.set_zlim([np.amin(landscape), np.amax(landscape)])
    ax.view_init(elev=56, azim=120)
    plt.rcParams.update({'axes.labelsize': 17})
    plt.show()

interactive(plot_fractal_landscape, 
            power=IntSlider(value=4, min=1, max=9, step=1, description='Ітерації'), 
            roughness=FloatSlider(value=1.7, min=0.1, max=2.7, step=0.1, description='Грубість', readout=False), 
            cm=Dropdown(options=['gist_earth', 'terrain', 'brg', 'cubehelix', 'gist_stern', 'twilight','viridis', 'cividis'], value='terrain', description='Палітра'))
```

Розмір сітки має велике значення для цього процесу, оскільки він впливає на можливість рекурсивного застосування кроків алгоритму: $$2^n + 1$$

В процесі виконання алгоритму кожен крок розділяє квадрати та алмази на чотири менших. Починаючи з квадрата, ми хочемо, щоб кожен наступний рекурсивний крок міг застосовувати аналогічний поділ. Розмір $2^n + 1$ гарантує, що після кожного поділу у нас є центральна точка.

`Грубість` впливає на амплітуду випадкових змін — дещо відрегулювавши її та мапу кольорів можна отримати поверхню, нагадуючу океан.

```python
size = 257
ocean1 = diamond_square(size, 0.5, 239)
ocean2 = diamond_square(size, 0.2, 839)

fig = plt.figure(figsize=(10, 6))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
x, y = np.meshgrid(np.arange(size), np.arange(size))
ax1.plot_surface(x, y, ocean1, rstride=1, cstride=1, cmap='ocean')
ax1.axis('off')
ax1.view_init(elev=47, azim=37)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(x, y, ocean2, rstride=1, cstride=1, cmap='ocean')
ax2.axis('off')
ax2.view_init(elev=38, azim=60)

fig.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0, hspace=0)

plt.tight_layout()
plt.show()
```

Алгоритм "Diamond-Square" може бути адаптований і для створення 2D рельєфу ландшафтів. У цьому випадку, замість створення поверхні за допомогою сітки квадратів, ми створюємо лінію, яка імітує контур рельєфу, використовуючи подібний принцип додавання випадковості до середнього значення висот точок. 

```python
def midpoint_displacement(start, end, roughness, size, seed):
    points = np.linspace(start, end, num=size)
    points[1::2] = (start + end) / 2

    displacement = roughness

    iteration_steps = size - 1
    while iteration_steps > 1:
        half_step = iteration_steps // 2
        np.random.seed(seed)
        for i in range(0, size - 1, iteration_steps):
            mid_point_index = i + half_step
            avg_height = (points[i] + points[i + iteration_steps]) / 2
            displacement_adjustment = np.random.uniform(-displacement, displacement)
            points[mid_point_index] = avg_height + displacement_adjustment

        displacement *= np.power(2, -0.5)
        iteration_steps //= 2

    return points

def plot_landscape(power):
    colors = ['#0d9a6869', '#2e9a0d63', '#06791a45', '#02710163']
    fig, ax = plt.subplots(figsize=(9.6, 4.6))
    size = 2 ** power + 1
    roughness = [0.6, 0.4, 0.25, 0.13]
    starts = [2, 1.7, 1.4, 1]
    scapes=[]

    for idx, (color, roughnes, start) in enumerate(zip(colors, roughness, starts)):
        seed = idx * 177 
        landscape = midpoint_displacement(start, start, roughnes, size, seed)
        scapes.append(landscape)
        ax.plot(landscape, lw=0, zorder=3)
        ax.fill_between(range(size), landscape, color=color, alpha=0.4, zorder=3)
    
    #ax.axis('off')
    ax.set_xlim([0, size-1])
    ax.set_ylim([0, max(scapes[0]) + 0.2])     
    plt.rcParams.update({'axes.labelsize': 17})
    
    plt.show()
    
interactive(plot_landscape, power=IntSlider(min=0, max=10, step=1, value=4, description='Ітерації'))
```

Повертаючись до алгоритму для генерації 3D ландшафтів, його також використовують для генерації 2D зображень, нагадуючих хмари. В такій ситуації "третя координата" відповідає виключно за колір точки.

Фрактал нище відомий як "*фрактал хмари*" чи "*фрактал плазми*".

```python
def plot_fractal(power):
    size = 2 ** power + 1
    clouds = diamond_square(size, 10)
    
    fig, ax = plt.subplots(1, 2, figsize=(10, 10))

    im1=ax[0].imshow(clouds, cmap='Blues')

    im2=ax[1].imshow(clouds, cmap='nipy_spectral')

    for a in ax:
        a.axis('off')

    plt.show()
    
interactive(
    plot_fractal, 
    power=IntSlider(value=5, min=1, max=8, step=1, description='Ітерації')
)
```

### Берегові лінії
Алгоритм для моделювання берегових ліній нагадує собою алгоритм зміщення середньої точки, але на цей раз зміщення відбувається за випадково визначеним напрямком та величиною, що зменшується з кожною ітерацією. Процес починається з відносно простої фігури, такої як трикутник. Для кожної сторони цієї фігури визначається його середина, а потім вона зміщується на вектор, визначений функцією зміщення, яка в свою чергу забезпечує зменшення сміщення з кожною ітерацією. Процесс продовжується для кожної отриманної фігури. 

```python
triangle = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
rectangle = np.array([[0, 0], [0.5, 0], [0.5, 1], [0, 1]])
def displace_coastline(vertices, iterations, displacement_scale=0.19, seed=7):
    if iterations == 0:
        return vertices

    np.random.seed(seed)

    new_vertices = []
    for i in range(len(vertices)):
        midpoint = (vertices[i] + vertices[(i + 1) % len(vertices)]) / 2

        angle = np.random.rand() * 2 * np.pi
        displacement = displacement_scale * np.array([np.cos(angle), np.sin(angle)])
        displaced_midpoint = midpoint + displacement

        new_vertices.extend([vertices[i], displaced_midpoint])

    return displace_coastline(new_vertices, iterations - 1, displacement_scale * 0.64)

def get_coordinates(i, seed, arr):
    coastline = displace_coastline(arr, i, seed=seed)
    coastline = np.array(coastline)
    return np.append(coastline, [coastline[0]], axis=0)
    
def plot_coastline(iter):
    for i in range(iter + 1):
        coastline1 = get_coordinates(i, 98, triangle)
        coastline2 = get_coordinates(i, 4, rectangle)
    
    fig, ax = plt.subplots(2, 2, figsize=(7.5, 7.5))

    ax[0, 1].plot(coastline1[:, 0], coastline1[:, 1], marker='o', ms=6, mfc='red', mew=0)
    ax[0, 1].grid(True)

    ax[0, 0].plot(coastline1[:, 0], coastline1[:, 1], color='green', lw=0.05)
    ax[0, 0].axis('off')
    ax[0, 0].fill(coastline1[:, 0], coastline1[:, 1], color='green', alpha=0.3)

    ax[1, 1].plot(coastline2[:, 0], coastline2[:, 1], marker='o', ms=6, mfc='red', mew=0)
    ax[1, 1].grid(True)

    ax[1, 0].plot(coastline2[:, 0], coastline2[:, 1], color='brown', lw=0.05)
    ax[1, 0].axis('off')
    ax[1, 0].fill(coastline2[:, 0], coastline2[:, 1], color='brown', alpha=0.3)
    
    for row in ax:
        for col in row:
            col.axis('equal')
            
    plt.rcParams.update({'axes.labelsize': 17})
    plt.tight_layout()
    
    plt.show()

interactive(plot_coastline, iter=IntSlider(value=1, min=0, max=10, step=1, description='Ітерації'))
```

Концепція фрактальної розмірності була популяризована Бенуа Мандельбротом саме через аналіз берегових ліній. Берегова лінія є класичним прикладом фрактальної геометрії у природі. Її фрактальна розмірність зазвичай перевищує 1, але є меншою за 2.
